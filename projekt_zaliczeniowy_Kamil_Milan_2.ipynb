```python
#zadanie 1.6

def remove_stopwords(tekst):
  dok = nlp(tekst)
  result = ""
  for token in dok:
    if token.is_stop == False:
      result += token.text + " "
  return result     
df["text_without_stopwords"] = df["text"].apply(remove_stopwords)
```

```python
#zadanie 3.1

def find_person(tekst):
  doc = nlp(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "persName":
      result += ent.text + " "
  return result
df["persons"] = df["text"].apply(find_person)
```

```python
#zadanie 3.2

def find_place(tekst):
  doc = nlp(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "placeName":
      result += ent.text + " "
  return result
df["places"] = df["text"].apply(find_place)
```

```python
#zadanie 3.3

def find_org(tekst):
  doc = nlp(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "orgName":
      result += ent.text + " "
  return result
df["organizations"] = df["text"].apply(find_org)
```

```python
#zadanie 3.4

nlp_blank = spacy.blank("pl")
ruler = nlp_blank.add_pipe("entity_ruler")
pattern = "\S+"
patterns = [{"label": "HASHTAG", "pattern": [{"ORTH": "#"}, {"TEXT": {"REGEX": pattern}}]}]
ruler.add_patterns(patterns)
def find_hashtag(tekst):
  doc = nlp_blank(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "HASHTAG":
      result += ent.text + " "
  return result
df["hashtags"] = df["text"].apply(find_hashtag)
```
